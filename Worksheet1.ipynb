• For the following arrays:

Prove following with Numpy:
1. Prove A.A−1 = I.
2. Prove AB ̸= BA.

.
• Solve the following system of Linear equation using Inverse Methods.
2x − 3y + z = −1
x − y + 2z = −3
3x + y − z = 9
{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” function
of Numpy}

```python
import numpy as np
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

#prove:A*A^-1=I
A1 = np.linalg.inv(A)
I = A.dot(A1)
I = np.round(I)
print("A*A^-1=I:\n",I)

#Prove: AB != BA
AB = A.dot(B)
BA = B.dot(A)
print("AB:\n",AB)
print("BA:\n",BA)
print("AB != BA")

#Prove (AB)^T = B^T A^T
left_side = AB.T
right_side = B.T.dot(A.T)
print(("(AB)^T:\n"),left_side)
print("B^T A^T:\n",right_side)
print("(AB)^T = B^T A^T")

#Linear equation using inverse method
A = np.array([[2,-3,1],
              [1,-1,2],
              [3,1,-1]])
B = np.array([-1,-3,9])

print(np.dot(np.linalg.inv(A),B))
```
